```python
import tensorflow as tf  
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
dataset = np.loadtxt("3D_print_quality_dataset.csv", dtype=float, delimiter="," )
x_train = dataset[:,:3]
y_train = dataset[:,3].astype(int)
print(y_train.shape)
```

```python
input_layer = keras.Input(shape=(3,),name="input_paramenters")
x = keras.layers.Dense(10,activation="relu")(input_layer)
x = keras.layers.Dense(10,activation="relu")(x)
x = keras.layers.Dense(10,activation="relu")(x)
net_out = keras.layers.Dense(1,activation="relu")(x)
net = keras.Model(input_layer,net_out,name="DeepNet")
net.summary()
opt = keras.optimizers.Adam(learning_rate=0.001,decay=1e-6)
```

```python
net.compile(opt,loss="mse")
net.fit(x_train,y_train, epochs=1000, batch_size=10,validation_split=0.1)
```

```python
x_test_indx = 27

# print(np.array([x_train[x_test_indx,:]]).T)
# print(x_train[x_test_indx,:].flatten())
example = net.predict([x_train[x_test_indx].reshape(1,3)])
print(example)
```
